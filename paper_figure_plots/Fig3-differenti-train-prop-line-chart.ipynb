```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3]

# 70%
y0 = [0.7067,0.683133333,0.833,0.9299]
y0err = [0.008418432,0.00306159,0.021360009,0.005071489]

# 50%
y1 = [0.7076,0.671366667,0.8344,0.928733333]
y1err = [0.011086929,0.012548439,0.025655604,0.005398457]

# 30%
y2 = [0.7011,0.656266667,0.8281,0.924266667]
y2err = [0.015878392,0.005717808,0.025402362,0.006274021]

# 10%
y3 = [0.6090,0.6105,0.799166667,0.905433333]
y3err = [0.0179,0.035973462,0.01886328,0.02069839]

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot0
ax.plot(x, y0, '-k', color='#AB88BE', label='100% Training Data')
ax.fill_between(x, y0 - np.array(y0err), y0 + np.array(y0err), alpha=0.2, facecolor='#AB88BE',
                linewidth=0.1, antialiased=True)

# Plot1
ax.plot(x, y1, '-k', color='#A8D3E3', label='70% Training Data')
ax.fill_between(x, y1 - np.array(y1err), y1 + np.array(y1err), alpha=0.2, facecolor='#A8D3E3',
                linewidth=0.1, antialiased=True)

# Plot2
ax.plot(x, y2, '-k', color='#ADD599', label='40% Training Data')
ax.fill_between(x, y2 - np.array(y2err), y2 + np.array(y2err), alpha=0.2, facecolor='#ADD599',
                linewidth=0.1, antialiased=True)

# Plot3
ax.plot(x, y3, '-k', color='#F6B37F', label='15% Training Data')
ax.fill_between(x, y3 - np.array(y3err), y3 + np.array(y3err), alpha=0.2, facecolor='#F6B37F',
                linewidth=0.1, antialiased=True)

# # Plot4
# ax.plot(x, y4, '-k', color='#AB88BE', label='Ours', linewidth=3)
# ax.fill_between(x, y4 - np.array(y4err), y4 + np.array(y4err), alpha=0.4, facecolor='#AB88BE',
#                 linewidth=0.1, antialiased=True)

ax.set_xticks(x)
ax.set_xticklabels(['Framingham', 'Breast', 'Stroke', 'Heart'], rotation=30, fontsize=15)
ax.set_ylabel('AUC', fontsize=18)


# plt.yticks(fontsize=15)

# plt.legend(fontsize=15)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='x', which='major', pad=7, length=10,width=2)
ax.tick_params(axis='y', which='major', pad=7, length=10, width=2)
# ax.spines['top'].set_linewidth(1.85)
# ax.spines['right'].set_linewidth(1.85)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)

ax.set_ylim(0.55, 1.0)
ax.set_yticks([0.6, 0.7, 0.8, 0.9, 1.0])
plt.savefig('ablation_study_comparison.pdf',bbox_inches = "tight",transparent=True)
```
