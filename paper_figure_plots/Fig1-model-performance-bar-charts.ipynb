# MIMIC-small performance in bar chart

```python
# Data
categories = ['Linear Regression', 'Random Forest', 'Support Vector Machine', 'Siamese Neural Network', 'Ours']
values = [0.61, 0.83, 0.51, 0.82, 0.95]
errors = [0.011, 0.0009, 0.009, 0.001, 0.012]

# Set the figure size
plt.figure(figsize=(15, 5))

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 7))

# Set the width of the bars
bar_width = 0.8
spacing    = 1.3                     # scale factor for the gaps
positions  = np.arange(len(categories)) * spacing

# Define a list of colors
colors = ['#DF9E9B', '#A8D3E3', '#ADD599', '#F6B37F', '#AB88BE']

# Create the bars with error bars
ax.bar(positions, values, yerr=errors, align='center', alpha=1.0, ecolor='grey', capsize=10, width=bar_width, color=colors)

# Set the x-axis tick labels to the category names and rotate them
ax.set_xticks(positions)
ax.set_xticklabels([])

# Set the y-axis label
ax.set_ylabel('AUC', fontsize=28)
# Increase the font size of the yticks
plt.yticks(fontsize=18)

ax.set_ylim([0.45, 1.0])

# Set the title
ax.set_title('', fontsize=28)

# Calculate the average value
average_value = np.mean(values[:-1])
plt.axhline(y=average_value, color='#4D4D4D', lw=3, linestyle='--', label=f'Average = {average_value:.2f}')

# plt.legend(fontsize=28)

# Display the plot
# plt.tight_layout()
# plt.show()

# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
ax.tick_params(axis='x', which='major', pad=7, length=5,width=1.85)
ax.tick_params(axis='y', which='major', pad=7, length=10, width=1.85)
ax.spines['top'].set_linewidth(1.85)
ax.spines['right'].set_linewidth(1.85)
ax.spines['left'].set_linewidth(1.85)
ax.spines['bottom'].set_linewidth(1.85)

plt.tight_layout()
plt.savefig('MIMIC_Dataset_S.pdf',bbox_inches = "tight",transparent=True)
```
