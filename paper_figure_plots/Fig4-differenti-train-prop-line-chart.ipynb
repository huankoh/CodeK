```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5,6]

# LR
y0 = [0.546,0.56,0.557,0.55,0.58,0.57,0.593]
y0err = [0.042,0.072,0.065,0.052,0.035,0.063,0.04]

# RF
y1 = [0.5217, 0.54, 0.547, 0.54, 0.523, 0.523, 0.54]
y1err = [0.064755412, 0.045825757, 0.073711148, 0.079372539, 0.005773503, 0.047258156, 0.052915026]

# SVM
y2 = [0.4752, 0.5, 0.52, 0.52, 0.513333333, 0.56, 0.573333333]
y2err = [0.0277, 0.065574385, 0.055677644, 0.06244998, 0.072341781, 0.045825757, 0.049328829]

# Ours
y4 = [0.573933333, 0.603633333, 0.592766667, 0.592533333, 0.6136, 0.6218, 0.658966667]
y4err = [0.012461675, 0.028080123, 0.019696785, 0.022753974, 0.02646356, 0.03570056, 0.044675534]

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(6, 3))

# Plot0
ax.plot(x, y0, '-k', color='#DF9E9B', label='Linear Regression')
ax.fill_between(x, y0 - np.array(y0err), y0 + np.array(y0err), alpha=0.2, facecolor='#DF9E9B',
                linewidth=5, antialiased=True)
ax.scatter(x, y0, color='#DF9E9B', edgecolors='white', zorder=3, s=100)

# Plot1
ax.plot(x, y1, '-k', color='#A8D3E3', label='Random Forest')
ax.fill_between(x, y1 - np.array(y1err), y1 + np.array(y1err), alpha=0.2, facecolor='#A8D3E3',
                linewidth=5, antialiased=True)
ax.scatter(x, y1, color='#A8D3E3', edgecolors='white', zorder=3, s=100)

# Plot2
ax.plot(x, y2, '-k', color='#ADD599', label='Support Vector Machine')
ax.fill_between(x, y2 - np.array(y2err), y2 + np.array(y2err), alpha=0.2, facecolor='#ADD599',
                linewidth=5, antialiased=True)
ax.scatter(x, y2, color='#ADD599', edgecolors='white', zorder=3, s=100)

# # Plot3
# ax.plot(x, y3, '-k', color='#F6B37F', label='Siamese Neural Network')
# ax.fill_between(x, y3 - np.array(y3err), y3 + np.array(y3err), alpha=0.2, facecolor='#F6B37F',
#                 linewidth=0.1, antialiased=True)

# Plot4
ax.plot(x, y4, '-k', color='#AB88BE', label='Ours', linewidth=3)
ax.fill_between(x, y4 - np.array(y4err), y4 + np.array(y4err), alpha=0.4, facecolor='#AB88BE',
                linewidth=5, antialiased=True)
ax.scatter(x, y4, color='#AB88BE', edgecolors='white', zorder=3, s=100)

ax.set_xticklabels(['', '10%', '20%', '30%', '40%', '50%', '60%', '70%'], rotation=0, fontsize=15)
ax.set_ylabel('AUC', fontsize=18)
plt.yticks(fontsize=15)

plt.legend(fontsize=15)


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='x', which='major', pad=7, length=10,width=2)
ax.tick_params(axis='y', which='major', pad=7, length=10, width=2)
# ax.spines['top'].set_linewidth(1.85)
# ax.spines['right'].set_linewidth(1.85)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)

ax.set_ylim(0.3, 0.7)
ax.set_yticks([0.3, 0.4, 0.5, 0.6, 0.7])

plt.tight_layout()
# plt.savefig('wgs_group1_results.pdf',bbox_inches = "tight",transparent=True)
plt.show()
```
